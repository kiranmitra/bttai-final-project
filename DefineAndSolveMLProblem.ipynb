**Data Prep**

Fewer features work better for KNNs so we will only use the most correlated features as well as including neighborhood information and room type information:
Entire home/apt, Hotel room, Private room, Shared room, Bronx, Brooklyn, Manhattan, Queens, Staten Island, minimum_nights, accommodates, bathrooms, availability_60, maximum_maximum_nights

```python
y = df['label_price']
X = df[['Entire home/apt', 'Hotel room', 'Private room', 'Shared room', 'Bronx', 'Brooklyn', 'Manhattan',
        'Queens', 'Staten Island', 'minimum_nights', 'accommodates', 'bathrooms', 'availability_60', 
        'maximum_maximum_nights']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**KNN**

```python
print('Running Grid Search...')
num_examples = X_train.shape[0]
param_grid = {'n_neighbors': [n for n in range(2,int(num_examples**0.5))]}

model = KNeighborsRegressor()

#Run a grid search with 5-fold cross-validation and assign the output to the object 'grid'.
grid = GridSearchCV(model, param_grid, cv=5)


#Fit the model (use the 'grid' variable) on the training data and assign the fitted model to the 
#    variable 'grid_search'

# YOUR CODE HERE
grid_search = grid.fit(X_train, y_train)

print('Done')
```

```python
best_k = grid_search.best_estimator_.n_neighbors
print("Best K: " + str(best_k))
```

```python
# Initialize the model with best_k
knn_model = KNeighborsRegressor(n_neighbors=best_k)
    
# Train the model using the training sets
knn_model.fit(X_train, y_train) 
    
# Make predictions on the test set
predictions = knn_model.predict(X_test) 
    
# Compute and print model's metrics
knn_r2 = r2_score(y_test,predictions)
print('R2 of KNN model: ' + str(knn_r2))
knn_mae = mae(y_test,predictions)
print('MAE of KNN model: ' + str(knn_mae))
knn_rmse = rmse(y_test,predictions)
print('RMSE of KNN model: ' + str(knn_rmse))
```

**Linear Regression**

```python
y = df['label_price']
X = df[['Entire home/apt', 'Hotel room', 'Private room', 'Shared room', 'Bronx', 'Brooklyn', 'Manhattan',
        'Queens', 'Staten Island', 'minimum_nights', 'accommodates', 'bathrooms', 'availability_60', 
        'maximum_maximum_nights']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Create the  LinearRegression model object 
linear_model = LinearRegression()

# Fit the model to the training data 
linear_model.fit(X_train, y_train)

#  Make predictions on the test data 
predictions = linear_model.predict(X_test)

# Compute and print model's metrics
linear_r2 = r2_score(y_test,predictions)
print('R2 of Linear model: ' + str(linear_r2))
linear_mae = mae(y_test,predictions)
print('MAE of Linear model: ' + str(linear_mae))
linear_rmse = rmse(y_test,predictions)
print('RMSE of Linear model: ' + str(linear_rmse))
```

Better scores than KNN!

**Data Prep**

Fewer features work better for Decision Trees, so we will only use the most correlated features as well as including neighborhood information and room type information: Entire home/apt, Hotel room, Private room, Shared room, Bronx, Brooklyn, Manhattan, Queens, Staten Island, minimum_nights, accommodates, bathrooms, availability_60, maximum_maximum_nights

```python
y = df['label_price']
X = df[['Entire home/apt', 'Hotel room', 'Private room', 'Shared room', 'Bronx', 'Brooklyn', 'Manhattan',
        'Queens', 'Staten Island', 'minimum_nights', 'accommodates', 'bathrooms', 'availability_60', 
        'maximum_maximum_nights']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**Decision Tree**

```python
print('Running Grid Search...')

# Create a range of hyperparameter values for 'max_depth'. 
hyperparams_depth = [2**n for n in range(2,5)]

# Create a range of hyperparameter values for 'min_samples_leaf'. 
hyperparams_leaf = [25*2**n for n in range(0,3)]

# Create parameter grid.
param_grid={'max_depth':hyperparams_depth, 'min_samples_leaf':hyperparams_leaf}

#Initialize Model
model = DecisionTreeRegressor()
 
# Run a Grid Search with 5-fold cross-validation using the model
grid = GridSearchCV(model, param_grid, cv=5)

# Fit the model on the training data
grid_search = grid.fit(X_train, y_train)

print('Done')
```

```python
print(grid_search.best_params_)
```

```python
# Initialize the model using best params
DT_model = DecisionTreeRegressor(max_depth = 16, min_samples_leaf = 25)
        
#Fit model to training data
DT_model.fit(X_train, y_train)
        
#Make predictions on the test data 
predictions = DT_model.predict(X_test)
        
# Compute and print model's metrics
dt_r2 = r2_score(y_test,predictions)
print('R2 of DT model: ' + str(dt_r2))
dt_mae = mae(y_test,predictions)
print('MAE of DT model: ' + str(dt_mae))
dt_rmse = rmse(y_test,predictions)
print('RMSE of DT model: ' + str(dt_rmse))
```

Decision Tree provided better results than KNN and Linear Regression, but we could do even better. 

**Random Forest**

```python
y = df['label_price']
X = df.drop(columns='label_price')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Create random forest with 20 trees
rf_model = RandomForestRegressor(n_estimators=20)

# Fit the model to the training data
rf_model.fit(X_train, y_train)

# Make predictions on the test data
rf_preds = rf_model.predict(X_test)

# Compute and print model's metrics
rf_r2 = r2_score(y_test,rf_preds)
print('R2 of RF model: ' + str(rf_r2))
rf_mae = mae(y_test,rf_preds)
print('MAE of RF model: ' + str(rf_mae))
rf_rmse = rmse(y_test,rf_preds)
print('RMSE of RF model: ' + str(rf_rmse))
```

Random forest is much better! I investigated other n_estimators and the performance did not change. Even using the best parameters found for the decision tree above resulted in the same performance. Also tried using less features but then performance worsened.

**Gradient Boosted Decision Trees**

```python
y = df['label_price']
X = df.drop(columns='label_price')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Initialize the model
gbdt_model = GradientBoostingRegressor(n_estimators = 100, max_depth = 16, min_samples_leaf = 25)

# Fit the model to the training data 
gbdt_model.fit(X_train, y_train)

# Make predictions on the test data
gbdt_preds = gbdt_model.predict(X_test)

# Compute and print model's metrics
gbdt_r2 = r2_score(y_test,gbdt_preds)
print('R2 of GBDT model: ' + str(gbdt_r2))
gbdt_mae = mae(y_test,gbdt_preds)
print('MAE of GBDT model: ' + str(gbdt_mae))
gbdt_rmse = rmse(y_test,gbdt_preds)
print('RMSE of GBDT model: ' + str(gbdt_rmse))
```

GBDT has the best results so far (lowest RMSE, lowest MAE, and highest R2)! This is the chosen best model.
